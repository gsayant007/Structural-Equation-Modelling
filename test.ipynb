```python
df = pd.read_csv("./dummy_data.csv")
df.head()
```

```python
# scale the data for better model fitting ? (not sure about this)
X = MinMaxScaler().fit_transform(df)
X = pd.DataFrame(X,columns=df.columns)
```

```python
desc3 = """
et0 =~ G
eta1 =~ y1
eta2 =~ y2+y3+y4
y1 ~~ y2+y3+y4
eta2 ~ x1+x2+x3
eta1 ~ eta2
et0 ~ eta1+eta2
z2 ~ eta2
G ~ z2
z2 ~ nx1+nx2+nx3+nx4+nx5+nx6+nx7+nx8+nx9+nx10+nx11+nx12+nx13
nx_lat2 =~ nx14+nx15
et0 ~ nx_lat2
nx_lat3 =~ nx16+nx17
et0 ~ nx_lat3
"""
```

```python
# create the model and fit the scaled data and see the result
mod = semopy.Model(desc3)
res = mod.fit(X)
print(res)
```

```python
# Finally let's inspect the parameters estimates
ins = mod.inspect()
print(ins)
```

```python
# visualize the result
g = semopy.semplot(mod,"pd1.png")
print(g)
```
